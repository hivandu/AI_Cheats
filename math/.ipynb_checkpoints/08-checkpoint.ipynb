# 导数

1. 迭代法求解二次函数最小值

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

```python
# 定义二次函数
def f(x):
    y = x**2
    return y
```

```python
# 定义对应的导数
def d(x):
    dy = 2*x
    return dy
```

```python
# 操作者输入指定的x初始值
x = float(input('请输入x的初始值：'))

# 定义学习率
lr = 0.2

# 准备一段二次函数数据点
x_temp = np.linspace(-x, x, 1000)
y_temp = x_temp**2

while abs(x) > 0.01:
    # 用于清空之前的输出
    display.clear_output(wait = True) 
    # 清除之前的列表
    plt.clf()
    plt.plot(x_temp, y_temp)
    plt.scatter(x, f(x), s=30, c='red')
    plt.show()
    print('此时x的值为: ', x)
    plt.pause(0.7)
    %matplotlib inline
    dy = d(x)
    x -= lr * dy
else:
    display.clear_output(wait=True)
    plt.clf()
    plt.plot(x_temp, y_temp)
    plt.scatter(x, f(x), s=30, c='red')
    plt.show()
    print('最终x的值为：', x)
```
