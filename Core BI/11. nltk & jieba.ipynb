# 词云

```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
```

```python
# 去掉停用词
def remove_stop_words(f):
    stop_words = ['Movie']
    for stop_word in stop_words:
        f = f.replace(stop_word, '')
    return f

# 生成词云
def create_word_cloud(f):
    print('根据词频，开始生成词云！')
    # 去掉停用词（高频词，比如the、of、a）
    f = remove_stop_words(f)    
    cut_text = word_tokenize(f)

    # print(cut_text)

    cut_text = ' '.join(cut_text)
    wc = WordCloud(
        max_words=100,
        width=1200,
        height=1000
    )
    wordcloud = wc.generate(cut_text)
    # 写词云图片
    wordcloud.to_file('assets/wordcloud.jpg')
    # 显示词云文件
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
# 数据加载
data = pd.read_csv('/Volumes/Media Database/SynologyDrive/data/AI_Cheats/wordcloud/movies.csv')
data.head()
```

```python
# 读取title和genres字段
title = ' '.join(data['title'])
genres = ' '.join(data['genres'])
all_word = title + genres
# 生成词云
create_word_cloud(all_word)
```

## Market_Basket_Optimisation

```python
# 数据加载
data = pd.read_csv('/Volumes/Media Database/SynologyDrive/data/AI_Cheats/wordcloud/Market_Basket_Optimisation.csv', header=None)
data.head()
```
