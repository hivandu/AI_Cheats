```python
# 设置转移矩阵
a = np.array([[0, 1/2, 1, 0], 
              [1/3, 0, 0, 1/2], 
              [1/3, 0, 0, 1/2], 
              [1/3, 1/2, 0, 0]])

a_leak = np.array([[0, 0, 0, 1/2], 
                   [0, 0, 0, 1/2], 
                   [0, 1, 0, 0], 
                   [0, 0, 1, 0]])

a_sink = np.array([[0, 0, 0, 0], 
                   [1/2, 0, 0, 1], 
                   [0, 1, 1, 0], 
                   [1/2, 0, 0, 0]])

b = np.array([1/4, 1/4, 1/4, 1/4])

w = b
```

```python
def work(a, w):
  for i in range(100):
    # 每次求解 w, 是向量
    w = np.dot(a, w)
    print(w)

# n 代表网络中的 page 个数
def random_work(a, w, n):
  d = 0.85
  for i in range(100):
    # 随机浏览模型,让节点每次不会等于 0
    w = (1-d)/n + d*np.dot(a, w)
    print(w)

# work(a, w)
# work(a_sink, w)
# work(a_leak, w)
# random_work(a, w, 4)
random_work(a_leak, w, 4)
# random_work(a_sink, w, 4)
```
