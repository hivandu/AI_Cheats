```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(5, size=(10, 2)), index = pd.date_range('1/1/2023', periods=10), columns=['A', 'B'])

# 将 B 列最近 2 个值相加并生成新列
df['C'] = df['B'].rolling(window=2).sum()
df['D'] = df['B'].rolling(window=5).sum()

# 将 B 列最近 2 个值求平均并生成新列
df['E'] = df['B'].rolling(window=2, min_periods=1).mean()
df['F'] = df['B'].rolling(window=5, min_periods=1).mean()
df
```
