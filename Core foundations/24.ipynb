# 深度学习进阶 1

```python
import numpy as np
import torch
from torch import nn
import torch.functional as F
```

```python
x = torch.from_numpy(np.random.random(size=(4, 10)))
```

```python
model = nn.Sequential(
    nn.Linear(in_features=10, out_features=5).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=5, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Sigmoid(),
    nn.Linear(in_features=8, out_features=8).double(),
    nn.Softmax()
)

ytrue = torch.randint(8, (4, ))
loss_fn = nn.CrossEntropyLoss()
```

```python
print(torch.randint(8, (4,)))
```

```python

print(model(x).shape)
```

```python
print(ytrue.shape)
```

```python
loss = loss_fn(model(x), ytrue)
```

```python
loss.backward()
```

```python
for p in model.parameters():
    print(p, p.grad)
```
