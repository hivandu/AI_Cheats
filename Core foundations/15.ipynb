## 支持向量机

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
label_a = np.random.normal(6, 2, size=(50, 2))
label_b = np.random.normal(-6, 2, size=(50, 2))
```

```python
label_a_x = label_a[:, 0]
label_b_x = label_b[:, 0]

w_and_b = []

for i in range(100):
    w, b = (np.random.random(size=(1, 2))*10-5)[0]
    w_and_b.append((w, b))
```

```python
def f(x, w, b):
    return w * x + b
```

```python
w_and_b_save = []
for w, b in w_and_b:
    if np.max(f(label_a_x, w, b)) <= -1 and np.min(f(label_b_x, w, b)) >= 1:
        w_and_b_save.append((w, b))

w_and_b_save
```

```python

plt.scatter(*zip(*label_a))
plt.scatter(*zip(*label_b))

for w, b in w_and_b_save:
    x = np.concatenate((label_a_x, label_b_x))
    plt.plot(x, f(x, w, b))

w, b = min(w_and_b, key = lambda w_b: w_b[0])
all_x = np.concatenate((label_a_x, label_b_x))
plt.plot(all_x, f(all_x, w, b), 'r-o')

plt.show()
```
