## 支持向量机 SVM

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
label_a = np.random.normal(6, 2, size=(50, 2))
```

```python
label_b = np.random.normal(-6, 2, size=(50, 2))
```

```python
plt.scatter(*zip(*label_a))
plt.scatter(*zip(*label_b))
```

```python
label_a_x = label_a[:, 0]
label_b_x = label_b[:, 0]
```

```python
def f(x, w, b):
    return w * x - b
```

```python
w_and_b = []
for i in range(100): 
    w, b = (np.random.random(size=(1, 2)) * 10 - 5)[0]
    if np.max(f(label_a_x, w, b)) <= -1 and np.min(f(label_b_x, w, b)) >= 1:
        print(w, b)
        w_and_b.append((w, b))
```

```python
plt.scatter(*zip(*label_a))
plt.scatter(*zip(*label_b))

for w, b in w_and_b:
    x = np.concatenate((label_a_x, label_b_x))
    plt.plot(x, f(x, w, b))
```

```python
plt.scatter(*zip(*label_a))
plt.scatter(*zip(*label_b))

w, b = sorted(w_and_b, key=lambda t: abs(t[0]))[0]
x = np.concatenate((label_a_x, label_b_x))
plt.plot(x, f(x, w, b))
```
